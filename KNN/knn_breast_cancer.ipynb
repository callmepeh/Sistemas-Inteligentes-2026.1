```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from ucimlrepo import fetch_ucirepo
```

```python
# Carregando o dataset
dataset = fetch_ucirepo(id=17)

x = dataset.data.features
y = dataset.data.targets.values

# Deixar em float
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
   
dataset.data.features.head()
```

```python
# Dividir o dataset em treino, teste e validação
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.3, random_state=42)
x_test, x_val, y_test, y_val = train_test_split(x_temp, y_temp, test_size=0.1, random_state=42)
```

```python
# Tratar dados médicos e normalização
x_train = np.nan_to_num(x_train, nan=0.0)
x_test = np.nan_to_num(x_test, nan=0.0)
x_val = np.nan_to_num(x_val, nan=0.0)


# Z-score
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_val = scaler.transform(x_val)
```

```python
# Distância entre validação e treino
distances = np.sqrt(np.sum((x_val[:, np.newaxis] - x_train) ** 2, axis=2))

# Achar melhor k com base em acurácia
k_values = range(1, 21)
best_k = 1
best_accuracy = 0

def most_common_label(labels):
    values, counts = np.unique(labels, return_counts=True)
    return values[np.argmax(counts)]

for k in k_values:
    # Encontrar os k vizinhos mais próximos
    neighbors = np.argsort(distances, axis=1)[:, :k]
    
    # Prever rótulos com base na maioria dos vizinhos
    y_pred = np.array([most_common_label(y_train[neighbors[i]].flatten()) for i in range(len(x_val))], dtype=object)
    
    # Calcular acurácia
    accuracy = accuracy_score(y_val.flatten(), y_pred)
    
    # Mostrar acurácia para cada k
    print(f'k={k}: Acurácia={accuracy:.4f}')
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_k = k

print(f'Melhor k: {best_k} com acurácia={best_accuracy:.4f}')
```

```python
# Distância teste e treino, usando o melhor k encontrado
distances_test = np.sqrt(np.sum((x_test[:, np.newaxis] - x_train) ** 2, axis=2))
nearest_indices_test = np.argsort(distances_test, axis=1)[:, :best_k]
nearest_distances_test = np.take_along_axis(distances_test, nearest_indices_test, axis=1)
predicted_labels_test = np.array([most_common_label(y_train[nearest_indices_test[i]].flatten()) for i in range(len(x_test))], dtype=object)

print(f'Acurácia no teste: {accuracy_score(y_test, predicted_labels_test):.4f}')
print('Matriz de Confusão:')
print(confusion_matrix(y_test, predicted_labels_test))
```

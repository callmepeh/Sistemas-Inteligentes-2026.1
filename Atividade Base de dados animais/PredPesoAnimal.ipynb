```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# 1. Pegamos TODAS as 15 variáveis originais
colunas_originais = [
    'AC', 'AG', 'CC', 'AP', 'P.C', 'CT', 'CO', 
    'CCAB', 'LIL', 'LIS', 'Cga', 'Cper', 'PerPe', 'Ccau', 'DC'
]

X_completo = df_final[colunas_originais].copy()
y = df_final['PESO'].copy()

# >>> A CORREÇÃO ENTRA AQUI: A FAXINA DAS VÍRGULAS <<<
# Transformamos todas as colunas de X_completo em string, trocamos ',' por '.' e convertemos para número
for coluna in X_completo.columns:
    if X_completo[coluna].dtype == 'object':  # Se o Pandas achou que é texto
        X_completo[coluna] = X_completo[coluna].astype(str).str.replace(',', '.').astype(float)

# Fazemos o mesmo para o y (PESO), só por garantia
if y.dtype == 'object':
    y = y.astype(str).str.replace(',', '.').astype(float)


# 2. A Mágica: Adicionamos os Cálculos Básicos
# Cálculo 1: Volume Corporal
X_completo['Volume_Corporal'] = X_completo['CT'] * X_completo['CC'] * X_completo['AC']

# Cálculo 2: Fórmula Zootécnica Clássica
X_completo['Formula_Zootecnica'] = (X_completo['P.C'] ** 2) * X_completo['CC']

# 3. Divisão Treino/Teste
X_train, X_test, y_train, y_test = train_test_split(X_completo, y, test_size=0.3, random_state=50)

# 4. Treinamento
modelo_rf = RandomForestRegressor(n_estimators=100, random_state=50)
modelo_rf.fit(X_train, y_train)

# 5. Avaliação
y_pred = modelo_rf.predict(X_test)
score_final = r2_score(y_test, y_pred)

print(f"R² com Todas as Variáveis + Fórmulas de Peso: {score_final:.4f}")
```
